### Occupation Popularity

From the dataset we want to see what the most popular occupations are in New Zealand rural area.

```python
occp = data[['OCCUPATION_M','house INCOME']].dropna()
occp.columns = ['Occupation','Income']
occp.head(10)
```

To know how many people there are in each Occupation/Income, we need to the dataset to be classified.<br/>
In order to compare and analyze quantity of people have different incomes and occupations, we convert above table into following form:

```python
occp[0]=0
#grouped = occp['OCCUPATION_M'].groupby(occp['house INCOME'])
grouped = occp.groupby(['Occupation','Income'])
convert_table = grouped.count()[0]
occp_new = convert_table.unstack()
occp_new
```

Draw a graph that shows population in each Occupation and Income.<br/>

(which are far more than other occupations)

And the occupations with least people are
* Machinery operator/driver
* Labourer

* Technician/trade worker

```python
occp_new[:12].plot.barh(stacked=True)
plt.title("Population of each occpation", alpha=1)
```

### Occupation & Income 
Above graph showed only population of each occpation in this survey. But it is not very clear to find the relationship between occupation and income, since there are different total of each occupation. So we need to find out the total and percentage instead of directly comparing population.

```python
#occp_new.ix[13] = occp_new.ix[:12].apply(np.sum, axis=0)
occp_new['Total']=occp_new.ix[:13].apply(np.sum, axis=1)
occp_new
```

For every occupation, we need to obtain it's ratio of each income range, in order to compare fairly. <br/>
The following function is used to transform the dataframe table and create a new table, changing all cells of value to percentages by dividing Total of each row.

```python
def convert_percentage(occp_percent):
    i=0
    for x in occp_percent.ix[:,0]:
        temp=[]
        for val in occp_percent.ix[i]:
            temp.append(round(val / occp_percent.ix[i].Total*100, 2))
        occp_percent.ix[i] = temp
        i=i+1
    return occp_percent

occp_percent = convert_percentage(occp_new).ix[0:,0:7]
occp_percent
```

```python
occp_percent.plot.bar()
plt.title("Percentage of Income Range For All Occupations ", alpha=1)
plt.ylabel('Percentage(%)')
```

Above graph partly represents the income ratio for each occupation in New Zealand rural area.

### Attitude To Survey Questions
What is most people's thought and attitude about these survey questions? 

```python
att = data.ix[0:,'ATT_Buy_NZ_productas_Strongly agree':'ATT_VOTEELEC_SD']
att.head(10)
```

The dataset is then transformed into below table, showing specific amount of people.

```python
att_new = att.count().to_frame('Count')
att_new['Strongly Agree']=0
att_new['Agree']=0
att_new['No Opinion']=0
att_new['Disagree']=0
att_new['Strongly Disagree']=0

count=att_new['Count']
att_new.drop('Count',axis=1, inplace=True)
i = 1
while i < 15:
    test = att_new.drop(att_new.index[i:i+4],axis=0, inplace=True)
    i=i+1

i = 0
while i < 14:
    att_new.ix[i] = count[i*5:i*5+5].values
    i=i+1
att_new.index=[u'Buy NZ Products',
       u'Environment Friendly Product', u'Quality',
       u'TECHIMP', u'FINCTL', u'WKBILLS',
       u'FINSEC', u'INVRISK', u'HMIDEA', u'SPEC',
       u'LOWFAT', u'HEALTHY', u'FASHION',
       u'VOTE ON LOCAL ISSUES']
att_new
```

Use the same approch to calculate total and percentage

```python
att_new['Total']=att_new.ix[:14].apply(np.sum, axis=1)
att_percent=convert_percentage(att_new).ix[:,:5]
att_percent
```

```python
att_percent.plot.bar()
plt.title("Percentage of Five Answers To Survey Questions ", alpha=1)
plt.xlabel('Questions')
plt.ylabel('Percentage(%)')
```
